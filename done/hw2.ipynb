```python
t = Table.from_file("data/smoking", data_start="###")
plt.scatter(t['LUNG'],t['CIG'], marker='.')
plt.title('Lung cancer deaths against Cigarette sales')
plt.xlabel('Deathes per 100K population from lung cancer')
plt.ylabel('Number of cigarettes smoked (hds per capita)')
for label, x, y in zip(t['STATE'], t['LUNG'],t['CIG']):
    plt.annotate(label,(x, y))
plt.show()

def normalize(x):
    s,m = np.std(x), np.mean(x)
    return [(a - m)/s for a in x]

def corr(x,y):
    return round((sum(map(o.mul, normalize(x), normalize(y))) / len(x)),3)

def remove_outliers(t, label):
    q1,q3 = np.percentile(t[label], 25.0), np.percentile(t[label], 75.0)
    max_val, min_val = q3 + 1.5 * (q3 - q1), q1 - 1.5 * (q3 - q1)
    return t.filter_rows(lambda row: min_val < row[label] < max_val)

t2 = remove_outliers(t, 'CIG')
print "B) cigarette sales vs lung cancer with, w/o outliers: ", corr(t['LUNG'],t['CIG']), corr(t2['LUNG'],t2['CIG'])
print "Removing outliers makes cigarette sales and lung cancer more related, causing one to reasonably guess that increasing sales will increase lung cancer. "
print
print "C) cigarette sales vs bladder cancer with, w/o outliers: ", corr(t['CIG'],t['BLAD']), corr(t2['CIG'],t2['BLAD'])
print "Removing outliers makes cigarette sales and bladder cancer less related, perhaps because this is a deasese of old folks and smoking kills you before you get old."
print

print "D) cigarette sales vs kidney cancer with, w/o outliers: ", corr(t['CIG'],t['KID']), corr(t2['CIG'],t2['KID'])
print "Removing outliers makes cigarette sales and kidney cancer more related, causing one to reasonably guess that increasing sales will increase kidney cancer. "
print

print "E) cigarette sales vs leukemia with, w/o outliers: ", corr(t['CIG'],t['LEUK']), corr(t2['CIG'],t2['LEUK'])
print "Removing outliers makes cigarette sales and leukemia more related, causing one to reasonably guess that increasing sales will decrease leukemia. "
print
print "F,G) We can't tell causation from this data. For leukemia, there is very low correlation so it's probabaly just noisy data"
```
